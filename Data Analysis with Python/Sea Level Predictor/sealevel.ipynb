```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress as res
```

```python
df=pd.read_csv('epa-sea-level.csv')
```

```python
df
```

```python
plt.scatter(df['Year'],df['CSIRO Adjusted Sea Level'],label='original data')
```

```python
slope, intercept, r, p, se = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
```

```python
plt.figure(figsize=(20, 6))
plt.scatter(df['Year'],df['CSIRO Adjusted Sea Level'],label='original data')
x_line = np.arange(df['Year'].min(), 2051)
plt.plot(x_line, intercept + slope * x_line, color='r', label='fitted line')
plt.legend()
plt.show()
```

```python
df_adjusted=df[(df['Year'] >= 2000)]
slope, intercept, r, p, se = linregress(df_adjusted['Year'], df_adjusted['CSIRO Adjusted Sea Level'])
plt.figure(figsize=(20, 6))
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], label='original data')
x_line = np.arange(2000, 2051)
plt.plot(x_line, intercept + slope * x_line, color='g', label='fitted line')
```
